```python
# Initial Library Imports
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Import Data from Excel Spreadsheets 3 & 4
summary_path = Path('dataset/EPL_dfp_summary.xlsx')
summary_2020 = pd.read_excel(summary_path, sheet_name=2)
summary_2019 = pd.read_excel(summary_path, sheet_name=3)
```

```python
# View Data imported
summary_2020.head()
```

```python
# Fix Typo and Set Index
summary_2020.rename(columns={'suhbscriber_id':'subscriber_id'}, inplace=True)
summary_2020.set_index('subscriber_id', inplace=True)
summary_2020.head()
```

```python
# Select only columns needed and rename columns
epl_2020 = summary_2020[['num_of_breaches', 'dfp_accounts']]
epl_2020.rename(columns={'num_of_breaches':'No_of_Breaches_2020', 'dfp_accounts':'DFP_Accounts_2020'}, inplace=True)
epl_2020.head()
```

```python
# View Data imported
summary_2019.head()
```

```python
# Fix Typo and Set Index
summary_2019.rename(columns={'suhbscriber_id':'subscriber_id'}, inplace=True)
summary_2019.set_index('subscriber_id', inplace=True)
summary_2019.head()
```

```python
# Select only columns needed and rename columns
epl_2019 = summary_2019[['num_of_breaches', 'dfp_accounts']]
epl_2019.rename(columns={'num_of_breaches':'No_of_Breaches_2019', 'dfp_accounts':'DFP_Accounts_2019'}, inplace=True)
epl_2019.head()
```

```python
# Concatenate the two years of data to one dataframe
epl_summary = pd.concat([epl_2019, epl_2020], axis=1, join='inner')
epl_summary.head()
```

```python
# Add the total number of Breaches
epl_summary['Total_No_of_Breaches'] = epl_summary['No_of_Breaches_2019'] + epl_summary['No_of_Breaches_2020']
epl_summary.head()
```

```python
# Add the total number of DFP Accounts
epl_summary['Total_DFP_Accounts'] = epl_summary['DFP_Accounts_2019'] + epl_summary['DFP_Accounts_2020']
epl_summary.head()
```

```python
# Import Data of Fraud Alerts CSV
fraud_path = Path('dataset/CM_Fraud_Alerts_20200429.csv')
fraud_alerts = pd.read_csv(fraud_path, delimiter='|')
fraud_alerts.head()
```

```python
# Drop Columns with useless data 
fraud_alerts.drop(columns=['subscriber_number', 'vendor', 'AlertId'], inplace=True)
fraud_alerts.head()
```

```python
# Check columns we're unsure of
fraud_alerts['bureau'].value_counts()
```

```python
# Check columns we're unsure of
fraud_alerts['alert_type'].value_counts()
```

```python
# Fix typos in alert_type column
fraud_alerts.loc[fraud_alerts['alert_type'] == 'Fraud  Alert', 'alert_type'] = 'Fraud Alert'
fraud_alerts.head()
```

```python
# Check column to make sure typo was changed
fraud_alerts['alert_type'].value_counts()
```

```python
fraud_alerts.dtypes
```

```python
# Change create_date to datetime
fraud_alerts['create_date'] = pd.to_datetime(fraud_alerts['create_date'], infer_datetime_format=True)
fraud_alerts.dtypes
```

```python
# Encode non-numerical columns
from sklearn.preprocessing import OneHotEncoder

# Create a list of categorical variables columns
categorical_variables = list(fraud_alerts.dtypes[fraud_alerts.dtypes == "object"].index)

# Display the categorical variables list
categorical_variables
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Encode the categorcal variables using OneHotEncoder
encoded_data = enc.fit_transform(fraud_alerts[categorical_variables])
```

```python
# Create a DataFrame with the encoded variables
encoded_df = pd.DataFrame(
    encoded_data,
    columns = enc.get_feature_names(categorical_variables)
)
# Review the DataFrame
encoded_df.head()
```

```python
# Add the encoded variables to the original DataFrame
encoded_fraud_alerts = pd.concat([encoded_df, fraud_alerts], axis=1)

# Review the Dataframe
encoded_fraud_alerts.head()
```

```python
# Set index of dataframe to subscriber_id, remove extraneous columns, and view
encoded_fraud_alerts.set_index('subscriber_id', inplace=True)
encoded_fraud_alerts.drop(columns=['bureau','alert_type'], inplace=True)
encoded_fraud_alerts.head()
```
