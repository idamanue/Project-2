```python
import pandas as pd
from pathlib import Path
```

```python
# Import data 

file_path = Path("dataset/EPL_dfp_summary.xlsx")
epl_summary_2020 = pd.read_excel(file_path, sheet_name= 2, index_col = "suhbscriber_id")
epl_summary_2019 = epl_summary2020 = pd.read_excel(file_path, sheet_name= 3, index_col = "suhbscriber_id")
```

```python
#view data imported 
epl_summary_2020.head()
```

```python
#view data imported 
epl_summary_2019.head()
```

```python
#select and rename columns needed 

epl_2020 = epl_summary_2020[['num_of_breaches', 'dfp_accounts']]
epl_2020.rename(columns = {'num_of_breaches': 'No of breaches 2020', 'dfp_accounts':'DFP accounts 2020'}, inplace= True) 
epl_2020.head()
```

```python
epl_2019 = epl_summary_2019[['num_of_breaches', 'dfp_accounts']]

epl_2019.rename(columns = {'num_of_breaches': 'No of breaches 2019', 'dfp_accounts':'DFP accounts 2019'}, inplace= True) 
epl_2019.head()
```

```python
epl_summary = pd.concat([epl_2019, epl_2020], axis=1, join='inner')
epl_summary.head()
```
